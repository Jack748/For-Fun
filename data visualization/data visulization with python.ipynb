```python
countries = pd.read_csv("countries.csv")
countries
```

```python
italy = countries[countries.country == "Italy"] 
spain = countries[countries.country == "Spain"]
print(italy)
i1 = italy.population.iloc[0]
s1 = spain.population.iloc[0]


plt.plot(italy.year, italy.population / i1 *100 -100)
plt.plot(spain.year, spain.population / s1 *100 -100)
plt.legend(["Italy", "Spain"])
plt.show()
```
